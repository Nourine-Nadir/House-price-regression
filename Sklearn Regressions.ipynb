```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data_array = []
csv_file_path='D:\\Artificial intelligence\\IA 2023 -\\Machine Learning\\Regression\\realEstate_price\\Real estate.csv'
# Open the CSV file
with open(csv_file_path, 'r') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:

        data_array.append(row)

data_array=np.array(data_array)
data=data_array[1:,:].astype(float)

df= pd.DataFrame(data_array)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[:,:-1])
output = data[:,-1:]
train_data,test_data,  train_label, test_label =train_test_split(data_scaled,output,test_size=0.2,random_state=12)

df.head()
```

## Linear Regression


```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,SGDRegressor
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import median_absolute_error

#----------------------------------------------------

linearModel = LinearRegression(fit_intercept=True,copy_X=True,n_jobs=-1)

linearModel.fit(train_data,train_label)

print('Linear Regression Train Score is : ' , linearModel.score(train_data, train_label))
print('Linear Regression Test Score is : ' , linearModel.score(test_data, test_label))
print('Linear Regression weights is : ' , linearModel.coef_)
print('Linear Regression bias is : ' , linearModel.intercept_)
y_pred= linearModel.predict(test_data)
print('Predicted Value for Linear Regression is : ' )
prediction= {
    'Y_pred':y_pred.ravel(),
    'Y_real':test_label.ravel()
}

#----------------------------------------------------

MAEValue = mean_absolute_error(test_label, y_pred, multioutput='uniform_average')
print('Mean Absolute Error Value is : ', MAEValue)

#----------------------------------------------------

MSEValue = mean_squared_error(test_label, y_pred, multioutput='uniform_average')
print('Mean Squared Error Value is : ', MSEValue)

#----------------------------------------------------

MdSEValue = median_absolute_error(test_label, y_pred)
print('Median Squared Error Value is : ', MdSEValue )

predictionDF=pd.DataFrame(prediction)
predictionDF.head(len(predictionDF))


```

## Quadratic Regression


```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)#Try degree=3 , you'll encounter overfitting (high trainScore , low TestScore)

newTrainData = poly.fit_transform(train_data)
newTestData = poly.fit_transform(test_data)

linearModel = LinearRegression(fit_intercept=True,copy_X=True,n_jobs=-1)

linearModel.fit(newTrainData,train_label)

print('Linear Regression Train Score is : ' , linearModel.score(newTrainData, train_label))
print('Linear Regression Test Score is : ' , linearModel.score(newTestData, test_label))
print('Linear Regression weights is : ' , linearModel.coef_)
print('Linear Regression bias is : ' , linearModel.intercept_)

y_pred= linearModel.predict(newTestData)
print('Predicted Value for Linear Regression is : ' )
prediction= {
    'Y_pred':y_pred.ravel(),
    'Y_real':test_label.ravel()
}

#----------------------------------------------------

MAEValue = mean_absolute_error(test_label, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Absolute Error Value is : ', MAEValue)

#----------------------------------------------------

MSEValue = mean_squared_error(test_label, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)

#----------------------------------------------------

MdSEValue = median_absolute_error(test_label, y_pred)
print('Median Squared Error Value is : ', MdSEValue )

predictionDF=pd.DataFrame(prediction)
predictionDF.head(len(predictionDF))
```

##### Notice that we have higher Train,Test Scores and lower MAE
