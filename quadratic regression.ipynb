```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
from sklearn.preprocessing import StandardScaler

```

```python
data_array = []
csv_file_path='C:\\My Github\\Machine Learning\\Regression\\realEstate_price\\Real estate.csv'
# Open the CSV file
with open(csv_file_path, 'r') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:

        data_array.append(row)

data_array=np.array(data_array) # transform to numpy array to apply its methods
data=data_array[1:,:].astype(float) # transform from string type to float

df= pd.DataFrame(data_array)
scaler = StandardScaler() 
data_scaled = scaler.fit_transform(data[:,:-1])# using here sklearn scaler to normalize data 

split_samples = np.floor(len(data_scaled)*0.8).astype(int) # split samples to 80% train 20% test

#train data
train_data = data_scaled[:split_samples,:]
train_label = data[:split_samples,-1:]
#Test data
test_data = data_scaled[split_samples:,:]
test_label = data[split_samples:,-1:]

```

```python
import numpy as np

class RL():
    def __init__(self, features, output) -> None:
        self.m, self.n = np.shape(features)
        self.weights = np.zeros((self.n, 1))  # Initialize weights as a 2D array
        self.bias = 0
        self.X = features
        self.outputs = output
        self.alpha = 0.01 
        self.regularization = 0.1

    def gradient_descent(self, epochs):
        W1 = self.weights
        W2 = self.weights
        b = self.bias
        alpha = self.alpha
        y = self.outputs
        X = self.X

        for e in range(epochs):
            y_pred = np.dot(np.power(X,2), W1) + np.dot(X, W2) + b

            dw1 = (1 / self.m) * np.dot(np.power(X.T,2), (y_pred - y))
            dw2 = (1 / self.m) * np.dot(X.T, (y_pred - y))
            db = (1 / self.m) * np.sum(y_pred - y)

            W1 = W1 - alpha * (dw1 + (self.regularization/self.m)*W1)
            W2 = W2 - alpha * (dw2 + (self.regularization/self.m)*W2)
            b = b - alpha * (db + (self.regularization/self.m)*b)

        return W1,W2, b
```

```python
features = train_data
array_output = train_label

cf = RL(features,array_output)

W1,W2,b=cf.gradient_descent(10000)
```

```python
print(f'weights {W1,b}')
res = np.dot(np.power(test_data,2),(W1)) +np.dot(test_data,(W2))+b

resultat={
    'predicted':res.ravel(),
    'output':test_label.ravel()
}
df_res=pd.DataFrame(resultat)




print('error',np.mean(abs(resultat['predicted']-resultat['output'])))
df_res.head(len(df_res))
```
