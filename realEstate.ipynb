```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
from sklearn.preprocessing import StandardScaler

```

```python
data_array = []
csv_file_path='D:\\My Github\\Regression\\realEstate_price\\Real estate.csv'
# Open the CSV file
with open(csv_file_path, 'r') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:

        data_array.append(row)

data_array=np.array(data_array)
data=data_array[1:,:].astype(float)

df= pd.DataFrame(data_array)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[:,:-1])

#Train data MODIFY IT TO 80 % INSTEAD ###############################################################################
train_data = data_scaled[:300,:]
train_label = data[:300,-1:]
#Test data
test_data = data_scaled[300:,:]
test_label = data[300:,-1:]

```

```python
import numpy as np

class RL():
    def __init__(self, features, output) -> None:
        self.m, self.n = np.shape(features)
        self.weights = np.zeros((self.n, 1))  # Initialize weights as a 2D array
        self.bias = 0
        self.X = features
        self.outputs = output
        self.alpha = 0.01

    def gradient_descent(self, epochs):
        W = self.weights
        b = self.bias
        alpha = self.alpha
        y = self.outputs
        X = self.X

        for e in range(epochs):
            y_pred = np.dot(X, W) + b

            dw = (1 / self.m) * np.dot(X.T, (y_pred - y))
            db = (1 / self.m) * np.sum(y_pred - y)

            W = W - alpha * dw
            b = b - alpha * db

        return W, b
```

```python
features = train_data
array_output = train_label
print(np.shape(features))
print(np.shape(array_output))

cf = RL(features,array_output)

W,b=cf.gradient_descent(15000)
```

```python
print(f'weights {W,b}')
res = np.dot(test_data,(W))+b

resultat={
    'predicted':res.ravel(),
    'output':test_label.ravel()
}
df_res=pd.DataFrame(resultat)




print('error',np.mean(abs(resultat['predicted']-resultat['output'])))
df_res.head(len(df_res))
```
